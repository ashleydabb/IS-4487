```python
df_decision = df[["leave"]].value_counts()
print(df_decision)
```

```python
fig, axis = plt.subplots()
axis.axis('equal')

minorcounts = df_decision

minorcounts.plot(kind='pie',autopct="%0.2f%%", labels=None, pctdistance=1.2, labeldistance=1)

axis.set_ylabel('')
axis.set_title('Decision')
axis.legend(loc=3, labels=df_decision.index)

plt.show()
```
